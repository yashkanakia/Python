## Question 1

```python
#Download IBM monthly data from Yahoo Finance. 

ibm_df = yf.download('IBM', start='2000-01-01', end='2004-12-31', progress=False)

#Estimate its standard deviation and Sharpe ratio from January 2000 to December 2004

print("Standard Deviation of IBM from January 2000 to December 2004: ", ibm_df['Adj Close'].std())

#Sharpe ratio is the ratio of average return divided by the standard deviation of returns annualized

#add a new column for daily return

ibm_df['Daily Return'] = ibm_df['Close'].diff()

sharpe_ratio = ibm_df['Daily Return'].mean()/ibm_df['Daily Return'].std()

print('Sharpe Ratio :', sharpe_ratio)
```

## Question 2

```python
def beta(X,y):
    return y.cov(X)/X.var()
```

```python
#What is the annual beta for IBM, DELL, and MSFT from 2017 to 2020
#Using ‘SPY’ represents the S&P 500 ETF Trust, which many people use as a surrogate for the total market

comp_data = yf.download(['IBM','DELL','MSFT','SPY'],start='2017-01-01', end='2020-12-31', progress=False)

comp_data.head()

df = comp_data['Adj Close']

df['DELL DR'] = df['DELL'].pct_change(periods=1)
df['IBM DR'] = df['IBM'].pct_change()
df['MSFT DR'] = df['MSFT'].pct_change()
df['SPY'] = df['SPY'].pct_change()

ibm_beta = beta(df['SPY'],df['IBM DR'])
dell_beta = beta(df['SPY'],df['DELL DR'])
msft_beta = beta(df['SPY'],df['MSFT DR'])


print("IBM Beta Coefficient: ", ibm_beta)
print("Dell Beta Coefficient: ", dell_beta)
print("MSFT Beta Coefficient: ", msft_beta)
```

<ul>Currently, you are working as a financial analyst at International Business Machine Corporation (IBM). The firm plans to issue 30-year corporate bonds with a total face value of $60 million in the United States. Each bond has a face value of $1,000. The annual coupon rate is 3.5%. The firm plans to pay coupons once every year at the end of each year. Answer the following three questions:
<li>How much would your company receive today by issuing the 30-year bonds?</li>
<li>What is the YTM (Yield to Maturity) of the bond?</li>
<li>How much extra money could your company receive if your company manages to increase its credit rating by one notch</li></ul>

```python
total_face_value = 60000000
face_value = 1000
coupon_rate = .035  
years = 30


#IBM rating is A+ & the spread over 30 years is 77 basis

rate_aa = 77/10000

#risk free 30 years interest rate is 3.55%

riskFree_rate = .0355

corp_ir_aa = rate_aa+riskFree_rate

unit_pv = npr.pv(corp_ir_aa,30,-coupon_rate*face_value,face_value)

no_of_bonds = total_face_value/face_value

totalpv = unit_pv*no_of_bonds
print("1. Company will receive today: ",totalpv)

irr = round(npr.rate(30,coupon_rate*face_value,-unit_pv,face_value)*100,3)
print("2. Yield to Maturity(YTM): ",irr,"%")

rate_aaplus = 89/10000
corp_ir_aaplus = rate_aaplus+riskFree_rate

unit_pvplus = npr.pv(corp_ir_aaplus,30,-coupon_rate*face_value,face_value)
totalpvplus = unit_pvplus*no_of_bonds

print("3. Extra Money Company will receive : ",totalpvplus-totalpv)
```
