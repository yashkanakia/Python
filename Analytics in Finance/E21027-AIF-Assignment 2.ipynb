```python
idf = pd.read_csv(r"C:\Users\Owner\Downloads\AIF\RawData\stock.csv")
idf.head().sort_values(by='Date')
```

```python
#function to normalize stock prices on their initial values

def normalize(df):
    x = df.copy()
    for i in x.columns[1:]:
        x[i] = x[i]/x[i][0]
    return x
```

```python
#normailze our stock data

idf_norm = normalize(idf)
```

```python
#initializes all the weights to 0
weights={}
for i in enumerate(idf.columns[1:]):
    weights[i[1]] = 0
```

```python
# as given in question the allocation is 50 % each for AAPL and AMZN, so assign the weights accordinglybb
weights['AAPL'] = .5
weights['AMZN'] = .5
```

```python
#initialize portfolio DF
portfolio_df = idf.copy()
for counter, stock in enumerate(idf.columns[1:]):
    portfolio_df[stock] = portfolio_df[stock]*weights[stock]
    #investment ammount is 1M - 1000000
    portfolio_df[stock] = portfolio_df[stock]*1000000

#Create a new column that contains sum of all the stock values
portfolio_df['Total daily worth'] = portfolio_df[portfolio_df !='Date'].sum(axis=1)
```

```python
#Calculate the daily retun

portfolio_df['Daily return'] = 0.0

for i in range(1,len(portfolio_df)):
    portfolio_df['Daily return'][i] = (portfolio_df['Total daily worth'][i] - portfolio_df['Total daily worth'][i-1])/portfolio_df['Total daily worth'][i-1]*100
    
portfolio_df.head()
```

```python
#plot the daily return

fig = px.line(x = portfolio_df['Date'],y = portfolio_df['Daily return'],title = 'Portfolio Daily % Return')
fig.show()
```
