### Hypotheses about the Difference in Two Sample Means

- Comparing Unemployment Rate in Wyoming County and Alabany County

#### H0: Sample Means are Equal 
#### H1: Sample Means are not Equal 

```python
a=np.var(df_NY['Age'])
b=np.var(df_TX['Age'])
print(a,b)
varianceValue = False
if a>b:
    c = a/b
elif b>a:
    c = b/a

if c<4:
    varianceValue = True

varianceValue
```

```python
tstat,pval = stats.ttest_ind(a= df_NY['Age'].values,
                b= df_TX['Age'].values,
                equal_var=True)
print(tstat,pval)
```

#### Confidence Interval of 95% 

```python
if pval <0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```
