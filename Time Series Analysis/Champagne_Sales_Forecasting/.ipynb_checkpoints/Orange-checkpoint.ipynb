```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima 
import warnings
warnings.filterwarnings("ignore")


# Read the Champagne dataset
df = pd.read_csv(r'C:\Users\Owner\Downloads\Miles_Traveled.csv',index_col='DATE')
df.head()
```

```python
#SARIMAX Library

import statsmodels.api as sm

mod = sm.tsa.statespace.SARIMAX(df['Miles_travelled'], trend='n', order=(1,0,1), seasonal_order=(1,0,1,12))
results = mod.fit()

# visualization
df['forecast'] = results.predict(start =280, end= 299, dynamic= True) 
plt.figure(figsize=(10,5))
plt.plot(df['Miles_travelled'], label='Train')
plt.plot(df['forecast'], label='Forecast')
plt.legend(loc='best')
plt.show()
```
